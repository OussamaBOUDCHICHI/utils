## Rappels : 

- Soit $\left(\mathcal{H}, \langle,\rangle \right)$ un espace de Hilbert (de dimension eventuellement infinie).

    - Une famille $(e_k)_{k\in I} \in \mathcal{H}^I$ est dite orthonormale ssi : $$\left(\forall \ i,j\in I\right):\hspace{0.5cm} \langle e_i, e_j \rangle = \delta_{i,j}$$

    - Elle est dite totale ssi : 
$$\overline{\mathrm{Vect}\left(e_i \ : \ i\in I\right)} = \mathcal{H}$$ 

    - Elle est dite hilbertienne ssi : elle est à la fois orthonormale et totale.

- Soit $(X, d)$ un espace métrique. Il est dit séparable si elle existe une suite dense dans $X$, i.e: il existe $(x_n)_{n\geq 0} \in X^{\mathbb{N}}$, tq : 
$$\left(\forall  \varepsilon > 0\right)  \left(\forall  x\in X\right)  \left(\exists  n_0\in \mathbb{N}\right):\hspace{0.5cm} d(x,x_{n_0}) \leq \varepsilon$$

- Soit $(\mathbf{\mathbb{R}^d}, \mu)$ l'espace mesurable $\mathbb{R}^d$  de **mesure positive bornée sur les ensemble boréliennes bornées**. On note l'espace des fonction intégrable d'ordre $p$ sur $X$ : $\mathcal{H}_p := L^p(X)$. Alors, l'ensemble des fonctions lisses à support compact $\mathcal{C}_c^{\infty}\left(\mathbb{R}^d\right)$ est dense dans $\mathcal{H}_p$ pour tout $p\geq 1$. Il s'ensuit qu'en particulier :

$$L^2(\mu) \text{ est séparable (et un espace de Hilbert)}$$

- $L^2(\mu)$ muni du p.s : 
$$\langle f,g\rangle := \int_{X}fgd\mu$$

est de Hilbert.

- Tout espace de Hilbert, **séparable** admet une base hilbertienne dénombrable.


- Ainsi : 
$$\color{blue}{\left(\exists (\nu_k)_{k\geq 0} \in L^2(\mu)^{\mathbb{N}} \text{ orthonormale} \right) \ \left(\forall \ f\in L^2(\mu) \right) \ \left(\exists (\beta_k)_{k\geq 0} \in\mathbb{R}^{\mathbb{N}}\right):\hspace{0.5cm} f = \sum_{k=0}^{+\infty} \beta_k \nu_k}$$
$$\color{darkred}{ \left(\forall \ k\in \mathbb{N}\right):\hspace{0.5cm} \beta_k = \langle f, \nu_k \rangle = \int_{X}f\nu_k d\mu}$$

- **Remarque** : Pour les démonstrations de ces théorèmes, le lecteur est renvoyé à n'importe quel livre d'analyse fonctionnelle.

## Introduction :

Soient $T > 0$ un horizon de temps fixé et $\mathbb{Q}$ est  l'unique probabilité risque neutre équivalente à la probabilité hitorique. (Sous les notations classiques) : On suppose que pour tout $\omega \in  \Omega$, la trajectoire $\tilde{\omega} : t\mapsto S_t(\omega)$ à valeurs dans $\mathbb{R}^d$ appartient à un certain espace fonctionnelle que l'on note $\mathbb{D}_d\left([0,T]\right) := \mathbb{D}_d\left([0,T], \mathbb{R}^d\right)$.

Soit $F_T : \mathbb{D}_d\left([0,T]\right) \mapsto \mathbb{R}^{l_T}$ une fonctionnelle mesurable (connue : typiqument il contient toute l'information relative à la trajectoire du processus $S$ qui nous sera utile dans le pricing de l'actif contingent en question). Soit $g:\mathbb{R}^{l_T} \mapsto \mathbb{R}$ une fonction borélienne (inconnue).

On note $X$ le payoff stochastique d'un actif contingent (normalement Path Dependent). $X$ est $\mathcal{F}_T$ mesurable. le prix de l'actif contingent est défini, à tout instant $t\in [0,T]$, par :
$$\mathbb{E}_{\mathbb{Q}}\left(X\exp\left(-\int_{t}^{T}r_sds\right) \ |\  \mathcal{F}_t\right)$$

On suppose que $\left(r_s\right)_s$ est constant à tout instant et on s'intéresse particulièrement à :
$$\mathbb{E}_{\mathbb{Q}}\left(X \ |\  \mathcal{F}_t\right)$$

On suppose que $X\in L^2(\mathbb{Q})$ et (comme on peut bien le démontrer) :
$$\mathbb{E}_{\mathbb{Q}}\left(X \ |\  \mathcal{F}_t\right) = g(F_t(S))$$

De ce fait : 
$$ g\in L^2\left(\mathbb{R}^{l_t}, \mathcal{B}(\mathbb{R}^{l_t}), \mathbb{Q}_{F_t(Z)}\right)$$

Notons $L^2_t := L^2\left(\mathbb{R}^{l_t}, \mathcal{B}(\mathbb{R}^{l_t}), \mathbb{Q}_{F_t(Z)}\right)$

On sait que $L_t^2$ muni du p.s : 
$$ \langle f, h\rangle_t = \int_{\mathbb{R}^{l_t}} f(x)g(x) d \mathbb{Q}_{F_t(Z)}(x) = \mathbb{E}_{\mathbb{Q}}\left[f\left(F_t(S)\right) h\left(F_t(S)\right)\right]$$

est un espace de hilbert séparable (une mesure de probabilité est toujours bornée est prositive). Elle admet donc (comme nous avons déjà mentionné) une base hilbertienne dénombrable et :
$$\left(\exists (\nu_k)_{k\geq 0} \in (L_t^2)^{\mathbb{N}} \text{ orthonormale} \right) \ \left(\exists (\beta_k)_{k\geq 0} \in\mathbb{R}^{\mathbb{N}}\right):\hspace{0.5cm} g = \sum_{k=0}^{+\infty} \beta_k \nu_k$$

tq : 
$$\left(\forall \ k\in \mathbb{N}\right):\hspace{0.5cm} \beta_k = \langle g, \nu_k \rangle = \mathbb{E}_{\mathbb{Q}}\left[g\left(F_t(S)\right) \nu_k\left(F_t(S)\right)\right] = \mathbb{E}_{\mathbb{Q}}\left[ \mathbb{E}_{\mathbb{Q}}\left(X \ |\  \mathcal{F}_t\right)\nu_k\left(F_t(S)\right)\right] = \mathbb{E}_{\mathbb{Q}}\left[ \mathbb{E}_{\mathbb{Q}}\left(X \nu_k\left(F_t(S)\right) |\  \mathcal{F}_t\right)\right] $$

Ainsi : 
$$\color{red}{ \left(\forall \ k\in \mathbb{N}\right):\hspace{0.5cm} \beta_k = \mathbb{E}_{\mathbb{Q}}\left(X \nu_k\left(F_t(S)\right)\right)}$$

On définit l'**erreur de** $\color{blue}{\textbf{projection}}$ : 
$$p_t(F_T(S)) := X - g(F_t(S))$$

La méthode LSMC essaye d'estimer la fonction $g$ à partir de son écriture dans la base hilbertienne ci-dessus. En utilisant des données simulées sous $\mathbb{Q}$.

Toutefois, en pratique, on ne peut pas simuler une infinité de données. Sur ce, on essaye de travailler sur une sous-famille finie de $\left(\nu_k\right)_k$, que l'on note $\left(\nu_k\right)_{1\leq k\leq K}$. 

Ainsi, on approxime $g$ par $g_K$ définie par : 
$$g_K := \sum_{k=1}^{K}\beta_k \nu_k$$

L'**erreur d'** $\color{blue}{\textbf{approximation}}$ est définie par : 
$$a_K := g - g_K$$

Remarque que : $\color{magenta}{a_K \underset{K\to +\infty}{\longrightarrow} 0}$. Et par orthogonalité de la base, on a par construction (donc):  
$$\color{magenta}{\langle g_K, a_K\rangle = \mathbb{E}_{\mathbb{Q}}\left[g_K\left(F_t(S)\right) a_K\left(F_t(S)\right)\right] = 0}$$

On peut maintenant écrire la régression suivante : 
$$X = g_K(F_t(S)) + a_K(F_t(S)) + p_t(F_T(S))$$

Maintenant, étant donné un échantillon simulé de taille $N$ : $\left(\left(x^j, A_t(S^j)\right)\right)_{1\leq j\leq N}$.  Il est naturel d'estimer $g_K$ par la projection : 
$$\hat{g}_K := \mbox{arg} \min_{g \in \mathcal{H}_K} N^{-1} \sum_{j=1}^{N}\left(x^j - g(A_t(S^j))\right)^2$$

tq : 
$$\mathcal{H}_K := \left\{ g:\mathbb{R}^{l_t} \mapsto \mathbb{R} \ : \left(\exists \ \left(\beta_1, \dots, \beta_K\right) \in \mathbb{R}^K\right),\ g = \sum_{k=1}^{K} \beta_k \nu_k\right\}$$

#### Les polynômes orthogonaux de Laguerre : 

Soit $n\in \mathbb{N}$, on définit les fonction $\psi_n$  et $L_n$ par : 
$$\left(\forall \ t\in \mathbb{R}\right):\hspace{0.5cm} \psi_n(t) = e^{-t} \frac{t^n}{n!}\hspace{0.5cm};\hspace{0.5cm} L_n(t) = \psi_n^{(n)}(t) e^t$$

La famille $\left(L_n\right)_n$ est une famille de polynôme de deg $n$ appelé : $\color{blue}{\textbf{polynômes de Laguerre}}$ (On confod ici, polynôme et fonction polynômiale).

On peut démontrer que : 
$$\left(\forall \ n\in \mathbb{N}\right):\hspace{0.5cm} L_n := \sum_{k=0}^{n}\mathrm{C}_n^k \frac{(-1)^k}{k!}X^k$$

```python
from scipy.special import binom     # binom(n,k) = C_n^k
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(10,8)
plt.style.use('seaborn-whitegrid')



def fact(n): 
    if (n==0): return 1
    else: return n*fact(n-1)
    
def L(x, n):
    ks = np.arange(n+1)   
    L = np.array([binom(n,k) * ((-1)**k / fact(k)) * (x**k) for k in ks]).sum()
    return L
    
     
```

```python
# Exemples : 

N = 10
Xs = np.linspace(10, 20, 100)
Ls = {}

for n, x in zip(range(1, N), Xs):
    Ls[n] = np.array([L(x, n) for x in Xs ])
```

```python
for n in range(1, N):
    plt.plot(Xs, Ls[n], linestyle='dashed', label='n = ' + str(n))
plt.legend()
```

- Utilisant ces polynômes, on désire construire une base orthormale de l'espace $L^2(\mathbb{R}_+)$. On définit alors la famille des fonctions $\left(\varphi_n\right)_n$ , par : 
$$\left(\forall \ t\in \mathbb{R}_+\right):\hspace{0.5cm} \varphi_n(t) := e^{-\frac{t}{2}}L_n(t)$$

On a bien $\left(\varphi_n\right)_n \in L^2(\mathbb{R}_+)^{\mathbb{N}}$ et : 
$$\left(\forall \ n,p \in \mathbb{N}\right):\hspace{0.5cm} \langle \varphi_n, \varphi_p\rangle = \int_{0}^{+\infty}L_n(t)L_p(t)e^{-t}dt  = \delta_{n,p}$$

La dernière égalité est obtenue par l'orthonormalité de la famille des polynômes de Laguerre par rapport aux produit scalaire défini sur $\mathbb{R}[X]$, par  : 
$$\langle  P, Q\rangle := \int_{0}^{+\infty}P(t)Q(t) e^{-t}dt$$


- On se place maintenant dans le cadre du modèle de Black-Scholes-Merton classique $(d=1)$. Condidèrons le problème du pricing d'une option asiatique arithmétique et discrète : 

    - $(t_j)_{0\leq j\leq K}$ une subdivision ordonnée de $[0,T]$, tel que : $t_0 = 0$ et $t_K = T$.
    
    -$$X := \left((1+K)^{-1} \sum_{j=0}^{K} S_{t_j} - K \right)^+$$

    -On a : $$\left(\forall \ \tilde{\omega} \in D_1\right):\hspace{0.5cm} A_{t_j}(\tilde{\omega}) := (j+1)^{-1} \sum_{i=0}^{j} \tilde{\omega}(t_i)$$

    - Maintenant, on observe que $l_{t_j} = 1$ et $g_{t_j} : \mathbb{R}+ \mapsto \mathbb{R}_{+}$
    
    - $$g_{t_j}^K  = \sum_{k=0}^{M} \beta_k \varphi_k$$
    
    - On peut calculer directement les $\beta_k$. Toutefois, leurs expressions ont l'air méchant.


```python
# Consider the pricing at t=0:
import scipy.stats as stats
import pandas as pd 

N = 5    # number of paths
T = 4    # Maturity
r = 0.02
K = 100
S_0 = 143
sigma = 0.2
```

```python
def B(N, T):
    dt = T / N
    Br = np.cumsum(np.sqrt(dt) * stats.norm.rvs(size=N))
    Br = np.insert(Br, 0, 0.0)
    return Br

def GBM(N,T, r, sigma, S_0):
    W = B(N, T)
    t = np.linspace(0, T, N+1)
    GB = S_0 * np.exp((r - (sigma**2)/2)*t + sigma*W)
    return t, GB
```

```python
S = {}

nbTradingsDays = 30 # number of (t_j)_j

for path in range(N+1):
    t, X = GBM(nbTradingsDays, T, r, sigma, S_0)
    S[path] = pd.Series(X, index=t)
```

```python
# rows : paths, columns : time
S = pd.concat(S, axis=1).T
```

```python
def Payoff(S, K):
    return np.maximum(S.mean() - K, 0)
```

```python
#Let's use general polynomials first: 


def LSMC_AmericanPut(nbOfPaths, nbOfTradingDays, r, K, T, S, deg=2):
    
    dt = T / (nbOfTradingDays) 
    df = np.exp(-r * dt)     # discount factor
    
    
    H = np.maximum(K - S, 0)
    
    '''
    H = np.zeros_like(S) 
    
    H[:,0] = np.maximum(S.iloc[:,0] - K,0)
    
    for j in range(1, S.shape[1]):
        for i in range(S.shape[0]):
            H[i,j] = Payoff(S.iloc[i, :j], K) 
    '''
    
    V = np.zeros_like(H)               
    V[:,-1] = H[:,-1]
    
    for t in range(N-2, 0, -1):
        validPaths = H[:, t] > 0
        
        rg = np.polyfit(S[validPaths, t], V[validPaths, t+1] * df, deg)
        C = np.polyval(rg, S[validPaths, t])
        
        #rg = np.polyfit(S.iloc[validPaths, t], V[validPaths, t+1] * df, deg)
        #C = np.polyval(rg, S.iloc[validPaths, t])
        
        exercise = np.zeros(len(validPaths), dtype=bool)     # initialize
        exercise[validPaths] = H[validPaths, t] > C          # paths where it is optimal to exercise

        V[exercise,t] = H[exercise,t]                        # set V equal to H where it is optimal to exercise 
        V[exercise,t+1:] = 0                                 # set future cash flows, for that path, equal to zero  
        discount_path = (V[:,t] == 0)                        # paths where we didn't exercise 
        V[discount_path,t] = V[discount_path,t+1] * df
    return V, df
   
```

```python
V, df= LSMC(N, nbTradingsDays, r, K, T, S)
```

```python
for j in range(1, S.shape[1]):
    print('at t = ', np.round(S.columns[j],2) , 'the price is : ', np.mean(V[ :j])*df) 
```
